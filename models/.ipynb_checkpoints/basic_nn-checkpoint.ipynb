```python
# Wujie Duan is responsible for this code
# credit to Professor Enric
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
from torch import nn
import torch.nn.functional as F
from torchsummary import summary

from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
# read csv files, uncomment to switch the dataset
# df = pd.read_csv('../dataset/train_with_countries.csv', index_col=0)
df = pd.read_csv('../dataset/train_no_countries.csv', index_col=0)
y_train = df['play_count']
df = df.drop(['play_count'], axis=1)
X_train = df

# df1 = pd.read_csv('../dataset/test_with_countries.csv', index_col=0)
df1 = pd.read_csv('../dataset/test_no_countries.csv', index_col=0)
y_val = df1['play_count']
df1 = df1.drop(['play_count'], axis=1)
X_val = df1

# show data
df1.head()
```

```python
# show columns
cols = df.columns.tolist()[:]
cols1 = df1.columns.tolist()[:]
```

```python
# MinMax Scaling
transformer = MinMaxScaler().fit(X_train) 
X_train = transformer.transform(X_train)
X_train = pd.DataFrame(X_train)

transformer = MinMaxScaler().fit(X_val) 
X_val = transformer.transform(X_val)
X_val = pd.DataFrame(X_val)

X_train.head()
```

```python
# show the shape of train and test set
print(X_train.shape)
print(X_val.shape)
```

```python
# set the size of train set
train_size = X_train.shape[0]
val_size = X_val.shape[0]

# convert to list
X_train = X_train.values.tolist()
y_train = y_train.values.tolist()
X_val = X_val.values.tolist()
y_val = y_val.values.tolist()
```

```python
# convert to tensor for pytorch
X_train = torch.tensor(X_train, dtype=torch.float, requires_grad=False)
y_train = torch.tensor(y_train, dtype=torch.float).reshape(train_size,1)
X_val = torch.tensor(X_val, dtype=torch.float, requires_grad=False)
y_val = torch.tensor(y_val, dtype=torch.float).reshape(val_size,1)
```

```python
# fit function for basic neural network
def fit(model, X_train, y_train, X_val, y_val, loss_fn, scheduler, optimizer, epochs = 1000, plotLosses=True):
    data   = []
    loss_dic = {"train": [], "validation": []}
    for i in range(epochs):
        
        # forward pass
        # use the scheduler
        scheduler.step()
        y_train_pred = model(X_train)

        # compute loss
        loss_train = loss_fn(y_train_pred, y_train)
        loss_dic["train"].append(loss_train / train_size)
        
        y_val_pred = model(X_val)
        
        loss_val = loss_fn(y_val_pred, y_val)
        loss_dic["validation"].append(loss_val / val_size)
        
        # Zero gradient, backward pass, gradient descent step
        optimizer.zero_grad()
        loss_train.backward()
        optimizer.step()
        
        # print losses for each epoch
        if i % 100 == 0:
            print(i, loss_train.item() / train_size, loss_val.item() / val_size)
    
    # plot losses
    if(plotLosses):
        plt.plot([x for x in range(epochs)], loss_dic['train'], label = "train loss")
        plt.plot([x for x in range(epochs)], loss_dic['validation'], label = "val loss")
        plt.xlabel("epoch")
        plt.ylabel("loss")
        plt.suptitle("train and val loss")
        plt.legend()
        plt.savefig('foo.png', bbox_inches='tight')
    return loss_dic
```

```python
# basic neural network model, credit to Professor Enric
# three layers: hidden layer 1 - hidden layer 2 - output layer
# input size: 126 (with countries) / 21 (no countries)
# output size: 1

class LocationNN(nn.Module):
    def __init__(self, input_size, hidden_size_1, hidden_size_2, output_size):
        super().__init__()
        self.hidden_layer_1 = nn.Linear(input_size, hidden_size_1)
        self.hidden_layer_2 = nn.Linear(hidden_size_1, hidden_size_2)
        self.output_layer = nn.Linear(hidden_size_2, output_size)
        
    def forward(self, x):
        x1 = F.relu(self.hidden_layer_1(x))
        x2 = F.relu(self.hidden_layer_2(x1))
        x3 = self.output_layer(x2)
        return x3
```

```python
# convert to float tensor
Xtorch_train  = torch.FloatTensor(X_train)
Ytorch_train  = torch.FloatTensor(y_train)
Xtorch_val  = torch.FloatTensor(X_val)
Ytorch_val  = torch.FloatTensor(y_val)
print(Ytorch_train.shape, Ytorch_val.shape)
```

```python
# set up parameters for basic nn
loc_model = LocationNN(
    input_size  = Xtorch_train.shape[1], 
    hidden_size_1 = 14,
    hidden_size_2 = 10,
    output_size = Ytorch_train.shape[1]
)
```

```python
# optimizer: Adam
optimizer = torch.optim.Adam(loc_model.parameters(), 3e-1)

# scheduler: StepLR
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.85, last_epoch=-1)

# loss function: L1 loss
lossfunc  = nn.L1Loss(reduction='sum')

# train the model
losses = fit(loc_model, Xtorch_train, Ytorch_train, Xtorch_val, Ytorch_val, lossfunc, scheduler, optimizer, epochs=100)

# save the model
torch.save(loc_model, "nn_no_model.tar")
# torch.save(loc_model, "nn_model.tar")

# show summary
summary(loc_model, (1, train_size, 21))

#show parameters
for i in loc_model.named_parameters():
    print(i)
```
